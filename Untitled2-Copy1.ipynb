This code tries to simulate a foucault pendulum 

```python
import numpy as np
from scipy.integrate import odeint
import math
from numpy import loadtxt
from matplotlib import pyplot as plt
import pandas as pd
from itertools import count
from matplotlib.animation import FuncAnimation
from IPython import display
```

```python
# Setting up the parameters
pi = 3.1416
g= 9.8    #m/s^2
theta = 0*2*pi/180 #radians
drag_coeff_linear = 0.7
drag_coeff_quad = 0.7
l = 7.98 #  ± 0.005    In the laboratory’s frame of reference, altitude of point of suspension
r = 6.371397 * pow(10,6)  # Radius of Earth (metres)
phi  = 0.788714289  #latitude (radians)
Omega_Earth = 7.2921150 * pow(10,-5)  #Earth’s angular rotational speed (rad/s)

omega = math.sqrt(g/l)
```

Now lets set up the system and create the parameter and the satte variable arrays/lists. We have to create four seperate 1sr order coupled diffrential equations to be passed into the RK4 method.


```python
A = 2.0*Omega_Earth*math.sin(phi)
B= 0.000
C= pow(omega,2)

P = np.array([A, B, C], dtype=np.float_)


def vectorfield(Y, t, P):
    """
    Defines the differential equations for the coupled system.

    Arguments:
        Y :  vector of the state variables:
                  Y = [x,y, v_x,v_y]
        t :  time
        P :  vector of the parameters:
                  P = [A,B,C]
    """
    x, y, u, v = Y
    A, B, C = P
    f = np.array([u, v, A * v - B * u - C * x, -A * u - B * v - C * y], dtype=np.float_)
    return f
```

To call in the RK4 method or built in odient method for integration. The code for the RK4 method isnt given here. I coded it speraterly and put in an external dependency that can be called upoon.

```python

abserr = 1.0e-6
relerr = 1.0e-6
stoptime = 24.0 * 60 * 60
numpoints = 25000
Y_0 = np.array([1,0 , 0, 0])
t = [stoptime * float(i) / (numpoints - 1) for i in range(numpoints)]
t=np.linspace(0,stoptime,numpoints)
sol = odeint(vectorfield, Y_0, t, args=(P,),
             atol=abserr, rtol=relerr)

```

Extracting the values from the solution and putting them into a dataframe.

```python
x = sol[:, 0]
y = sol[:, 1]
v_x = sol[:, 2]
v_y = sol[:, 3]
data =(t,x,y,v_x,v_y)
data = np.array(data)
df = pd.DataFrame(data.T,columns=['t','x','y','v_x','v_y'])
```

Setting up our plot. Initializing the figure and then poutting in the limits.

```python
fig = plt.figure()
lines = plt.plot([])
line = lines[0]
plt.xlim(-1,1)
plt.ylim(-1,1)
```

Defining the animate function. It will uypdate the plot regulary as iy reads in the new values for the variables. And drawing the animation

```python
def animate(i):
    #update plot
    line.set_data(x[:i],y[:i])

anim = FuncAnimation(fig,animate, frames=10000, interval=20)
video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)
plt.close()
```
